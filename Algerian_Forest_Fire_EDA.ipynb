```python
# Function to identify and treat outliers in the dataset.
def identifying_treating_outliers(df,col,remove_or_fill_with_quartile=None):
    """ This function will return the lower fence, upper fence and total outliers in the column of a dataframe and in the argument remove_or_fill_with_quartile you can pass drop to drop the outliers(rows) from the dataframe or you can pass fill to fill the lower fence value for outliers which are less than the lower fence and upper fence value for outliers which are greater the upper fence of that column."""
    q1=df[col].quantile(0.25)
    q3=df[col].quantile(0.75)
    iqr=q3-q1
    lower_fence=q1-1.5*(iqr)
    upper_fence=q3+1.5*(iqr)
    total_outliers=df[col].loc[df[col]<lower_fence].count() + df[col].loc[df[col]>upper_fence].count()
    if remove_or_fill_with_quartile=="drop":
        df.drop(df.loc[df[col]<lower_fence].index,inplace=True)
        df.drop(df.loc[df[col]>upper_fence].index,inplace=True)
    elif remove_or_fill_with_quartile=="fill":
        df[col] = np.where(df[col] < lower_fence, lower_fence, df[col]) # Replacing the outlier value with value of lower fence.
        df[col] = np.where(df[col] > upper_fence, upper_fence, df[col]) # Replacing the outlier value with value of upper fence.
    elif remove_or_fill_with_quartile==None:
        pass
    return lower_fence,upper_fence,total_outliers
```

```python
# Getting lower fence ,upper fence and total outliers for all numerical columns
for col in numerical_columns:
    print(f"For Column {col} lower fence is {np.round(identifying_treating_outliers(df,col)[0],2)}, upper fence is {np.round(identifying_treating_outliers(df,col)[1],2)}, and total outliers are {identifying_treating_outliers(df,col)[2]}")
```

```python
# Getting the region wise total cases of fire and not fire
df["Classes"].groupby(df["region"]).value_counts().to_frame()
```

```python
sns.countplot(df["Classes"],hue=df["region"])
plt.legend(bbox_to_anchor=(1.04, 1), loc="upper left") # Getting the legend outside the plot
```

```python
sns.scatterplot(x=df["RH"],y=df["Rain"],hue=df[['Classes', 'region']].apply(tuple, axis=1))
plt.legend(bbox_to_anchor=(1.04, 1), loc="upper left") # Getting the legend outside the plot
```

1. In Bejaia region humidity is  always above 50.
2. In Sidi-Bel Abbes region if the humidity is below 50 and there is no rain the forest will catch fire.
3. Rain plays an important role, as we can see even if there is little rain there is no sign of forest fire in both the regions( ignoring the outliers).

```python
sns.catplot(x="region",y="FWI",hue="Classes", kind="swarm",data=df)
plt.legend(bbox_to_anchor=(1.04, 1), loc="upper left") # Getting the legend outside the plot
```

If  Fire Weather Index (FWI) is more than 5 in both the regions forests will catch fire and if FWI is less than 5 there are only few cases of forest catching fire. This implies FWI plays an important role, this acts as an alarm, as soon as FWI crosses 3 units on FWI scale Forest Department should intervene and take all the necessary steps and make required arrangements.
